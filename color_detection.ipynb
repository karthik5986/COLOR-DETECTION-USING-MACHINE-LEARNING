```python
import cv2
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
img = cv2.imread("colorpic.jpg")
```

```python
data = pd.read_csv("colors.csv")
```

```python
clicked = False
r = g = b = xpos = ypos = 0
```

```python
data.head()
```

```python
# Dividing the dataset as dependent and Independent variables
x = data.iloc[:,3:6].values
y = data[['hex','color']].values

# Splitting the data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)

# Initializing the classifier
knn = KNeighborsClassifier(n_neighbors=1)

# Training the model
knn.fit(x_train,y_train)

# Testing the model
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.preprocessing import LabelEncoder
lf = LabelEncoder()
data['hexa'] = data['hex']
data['hex'] = lf.fit_transform(data['hex'])

data.head()
```

```python
x = data.iloc[:,3:6].values
y = data['hex'].values

model = Sequential()
model.add(Dense(12, input_shape=(x.shape[1],), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=150, batch_size=10)
```

```python
def draw_function(event, x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b,g,r,xpos,ypos, clicked
        clicked = True
        xpos = x
        ypos = y
        b,g,r = img[y,x]
        b = int(b)
        g = int(g)
        r = int(r)
def rgb_to_hex(rgb):
    return '#{:02x}{:02x}{:02x}'.format(rgb[0], rgb[1], rgb[2])
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_function)
while(1):

    cv2.imshow("image",img)
    if (clicked):
        out = round(model.predict([[r,g,b]])[0,0])
        color_name = data[data['hex']==out]['color'].values[0]

        cv2.rectangle(img,(20,20), (1000,60), (b,g,r), -1)

        cv2.rectangle(img,(20,20), (1000,60), (b,g,r), -1)


        text = "The model predicts as "  + color_name + ' R='+ str(r) +  ' G='+ str(g) +  ' B='+ str(b) + ' Hex=' + rgb_to_hex([r,g,b])


        cv2.putText(img, text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)


        if(r+g+b>=600):
            cv2.putText(img, text,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)

        clicked=False


    #Break the loop when user hits 'esc' key
    if cv2.waitKey(20) & 0xFF ==27:
        break

cv2.destroyAllWindows()
```
